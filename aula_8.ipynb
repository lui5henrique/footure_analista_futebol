```python
import random
import pandas as pd
import time
```

```python
# BREAK 

jogadores = ["Neymar Jr", "Vinicius Jr", "Lucas Paquet√°", "Alex Telles", "Igor Jesus",
            "Marquinhos", "Ederson", "Casemiro", "Bruno Guimaraes", "Raphinha", "Danilo",
            "Luiz Henrique", "Jo√£o Pedro", "Rodrygo", "Gabriel Magalhaes", "Cuiabano"]

# Simula o penalti (60% chance de gol)
for jogador in jogadores:
    fez_gol = random.choices([True, False])[0] # [0] sempre 0 por que tem apenas um elemento index na lista
    time.sleep(1)  # Pausa de 1 segundo para simular o tempo de chute
    
    if fez_gol:
        print(f"{jogador} marcou! Treino encerrado, v√° para o chuveiro.")
        break    
    else:
        print(f"{jogador} errou. Pr√≥ximo!")
```

```python
# CONTINUE

jogadores = ["Neymar Jr", "Vinicius Jr", "Lucas Paquet√°", "Alex Telles", "Igor Jesus",
            "Marquinhos", "Ederson", "Casemiro", "Bruno Guimaraes", "Raphinha", "Danilo",
             "Luiz Henrique", "Jo√£o Pedro", "Rodrygo", "Gabriel Magalhaes", "Cuiabano"]

for jogador in jogadores:
    esta_machucado = random.choices([True, False], weights= [50, 50])[0] # 10% chance de estar machucado
    time.sleep(0.5)  # Pausa de 0.5 segundo para simular o tempo de verifica√ß√£o
    
    if esta_machucado:
        print(f"{jogador} est√° machucado. Pulando...")
        continue    
        
    print(f"{jogador} fez o passe!")
```

```python
# LOOP ANINHADOS

jogadores = ["Messi", "D√≠ Maria", "Tiquinho Soares"]
dias_treino = ["Ter√ßa", "Sexta"]

for jogador in jogadores:
    for dia in dias_treino:
        time.sleep(0.5)  # Pausa de 0.5 segundo para simular o tempo de treino
        print(f"{jogador} treinou finaliza√ß√µes na {dia}.")
```

Exerc√≠cio 1 ‚Äì Usando break

Voc√™ √© o respons√°vel por convocar os capit√£es do time para uma reuni√£o com a comiss√£o t√©cnica.

Use um for para percorrer a lista de jogadores e imprimir o nome dos jogadores que s√£o capit√£es.

A reuni√£o s√≥ pode ter 3 capit√£es, ent√£o assim que encontrar os 3 primeiros, use o break para encerrar a busca.

```python
import time

jogadores = [
    {"nome": "Neymar Jr", "capitao": True},
    {"nome": "Vinicius Jr", "capitao": False},
    {"nome": "Lucas Paquet√°", "capitao": False},
    {"nome": "Alex Telles", "capitao": False},
    {"nome": "Igor Jesus", "capitao": False},
    {"nome": "Marquinhos", "capitao": True},
    {"nome": "Ederson", "capitao": False},
    {"nome": "Casemiro", "capitao": True},
    {"nome": "Bruno Guimaraes", "capitao": False},
    {"nome": "Raphinha", "capitao": False},
    {"nome": "Danilo", "capitao": True},
    {"nome": "Luiz Henrique", "capitao": False},
    {"nome": "Jo√£o Pedro", "capitao": False},
    {"nome": "Rodrygo", "capitao": False},
    {"nome": "Gabriel Magalhaes", "capitao": False},
    {"nome": "Cuiabano", "capitao": False}
]

contador_capitaes = 0

for jogador in jogadores:
    time.sleep(0.5)  # Pausa de 0.5 segundo para simular a verifica√ß√£o
    if jogador["capitao"]: 
        contador_capitaes += 1
        print(f'O jogador {jogador["nome"]} √© capit√£o do time.')
        if contador_capitaes == 3:  # Parou quando atingir 3 capit√£es
            print("Reuni√£o completa! 3 capit√£es convocados.")
            break
```

Exerc√≠cio 2 ‚Äì Usando break com cobran√ßas de falta

Simule um treino de cobran√ßa de falta. Para cada jogador da lista, gere aleatoriamente se ele fez ou perdeu o gol com probabilidade de 20% de chance de ser gol.

Assim que o primeiro jogador marcar, o treino deve ser encerrado com break.

```python
import random
import time

jogadores = ["Neymar Jr", "Vinicius Jr", "Lucas Paquet√°","Alex Telles", "Igor Jesus",
             "Marquinhos","Ederson","Casemiro", "Bruno Guimaraes","Raphinha","Danilo",
             "Luiz Henrique","Jo√£o Pedro","Rodrygo", "Gabriel Magalhaes","Cuiabano"]

for jogador in jogadores:
    gol_falta = random.choices([True, False], weights = [20, 80])[0]
    time.sleep(0.5)

    if gol_falta:
        print(f'O jogador {jogador}, fez o gol!')
        break
    else:
        print(f'O jogodador {jogador} perdeu o gol, tente outra vez!')
```

Exerc√≠cio 3 ‚Äì Convoca√ß√£o da Sele√ß√£o (com continue)

```python
import random
import time

jogadores = ["Neymar Jr", "Vinicius Jr", "Lucas Paquet√°","Alex Telles", "Igor Jesus",
             "Marquinhos","Ederson","Casemiro", "Bruno Guimaraes","Raphinha","Danilo",
             "Luiz Henrique","Jo√£o Pedro","Rodrygo", "Gabriel Magalhaes","Cuiabano"]

for jogador in jogadores:
    machucados = random.choices([True, False], weights = [15,75])[0]
    time.sleep(1)

    if machucados:
        continue
    else:
        print(f'{jogador.upper()} n√£o est√° machucado: Est√° convocado para a Copa 2026!')
```

Exerc√≠cio 5 ‚Äì Loop aninhado: treino + desempenho

Os jogadores far√£o treinos de finaliza√ß√£o em tr√™s dias da semana: segunda, quarta e sexta.

Crie um loop aninhado para simular os treinos de cada jogador, e dentro de cada treino, gere aleatoriamente quantos gols ele marcou (de 0 a 4 gols por treino).

Use random.randint(0, 4) para gerar os gols em cada treino.

Dica: o loop externo percorre a lista de jogadores, e o loop interno percorre os dias da semana.

```python
import random
import time

jogadores = ["Neymar Jr", "Vinicius Jr", "Lucas Paquet√°","Alex Telles", "Igor Jesus",
             "Marquinhos","Ederson","Casemiro", "Bruno Guimaraes","Raphinha","Danilo",
             "Luiz Henrique","Jo√£o Pedro","Rodrygo", "Gabriel Magalhaes","Cuiabano"]

dias_semana = ['Segunda', 'Quarta', 'Sexta']

for jogador in jogadores:
    for dia in dias_semana:
        time.sleep(0.5)
        gols = random.randint(0,4)
        print(f'{jogador.upper()} fez treinos de finaliza√ß√µes nos dias: {dia} e marcou {gols} gols.') 
```
